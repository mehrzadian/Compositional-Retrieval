```python
class RetrievalDataset_csv(Dataset):
    def __init__(self, csv_file, image_folder, transform, text_processor, max_text_length=77, negative_sample_prob=0.5):
        """
        Args:
            csv_file (str): Path to the CSV file containing query-target pairs and text.
            image_folder (str): Directory path where the images are stored.
            transform (callable): A function/transform to apply to the images (e.g., normalization).
            text_processor (transformer): Pre-trained transformer like CLIP processor or any other model tokenizer.
            max_text_length (int): Maximum length of the tokenized text (to avoid truncation).
            negative_sample_prob (float): Probability of returning a negative sample.
        """
        self.data = pd.read_csv(csv_file)
        self.image_folder = image_folder
        self.transform = transform
        self.text_processor = text_processor
        self.max_text_length = max_text_length
        self.negative_sample_prob = negative_sample_prob

    def __len__(self):
        """Returns the total number of samples in the dataset"""
        return len(self.data)

    def __getitem__(self, idx):
        """Fetches one sample of image-text pair from the dataset"""
        row = self.data.iloc[idx]

        # Load query and target images
        query_img_path = f"{self.image_folder}/{row['query_image']}"
        target_img_path = f"{self.image_folder}/{row['target_image']}"

        # Error handling for invalid image paths
        if not isinstance(query_img_path, str) or not query_img_path or not isinstance(target_img_path, str) or not target_img_path:
            print(f"Skipping invalid image path at index {idx}: query_img_path={query_img_path}, target_img_path={target_img_path}")
            return self.get_dummy_sample()  # Return a dummy sample in case of invalid data

        try:
            # Open images
            query_image = Image.open(query_img_path).convert("RGB")
            target_image = Image.open(target_img_path).convert("RGB")
        except FileNotFoundError:
            print(f"Image file not found at index {idx}: query_img_path={query_img_path}, target_img_path={target_img_path}")
            return self.get_dummy_sample()  # Return a dummy sample if images are missing

        # Apply transformations (resize, normalization, etc.)
        query_image = self.transform(query_image)
        target_image = self.transform(target_image)

        # Process text: Tokenize using the text processor (e.g., CLIP tokenizer)
        query_text = row['query_text']
        text_inputs = self.text_processor(text=query_text,
                                          padding=True,
                                          truncation=True,
                                          max_length=self.max_text_length,
                                          return_tensors="pt")

        # Optional: Sample a negative image based on probability
        negative_image = None
        if random.random() < self.negative_sample_prob:
            negative_image = self.get_negative_sample(idx)

        return query_image, target_image, text_inputs, negative_image

    def get_negative_sample(self, idx):
        """Sample a random image that is not the target of the current query"""
        negative_idx = random.randint(0, len(self.data) - 1)
        while negative_idx == idx:  # Ensure it's not the same index
            negative_idx = random.randint(0, len(self.data) - 1)

        row = self.data.iloc[negative_idx]

        negative_img_path = f"{self.image_folder}/{row['query_image']}"
        try:
            negative_image = Image.open(negative_img_path).convert("RGB")
            negative_image = self.transform(negative_image)
        except Exception as e:
            print(f"Error loading negative sample at index {negative_idx}: {e}")
            return None  # Handle any error loading the negative image

        return negative_image

    def get_dummy_sample(self):
        """Return a dummy sample in case of invalid data"""
        dummy_image = np.zeros((3, 224, 224), dtype=np.float32)  # A dummy black image
        dummy_text = {
            'input_ids': torch.tensor([[0]]),
            'attention_mask': torch.tensor([[0]])
        }
        return torch.tensor(dummy_image), torch.tensor(dummy_image), dummy_text, None
```

```python
class RetrievalDataset(Dataset):
    def __init__(self, dataset, transform=None, text_processor=None, max_text_length=77, negative_sample_prob=0.5):
        """
        Args:
            dataset (Dataset): A HuggingFace dataset (like CIRR, MSCOCO, or Flickr30k) with 'image', 'caption', and 'id' columns.
            transform (callable): A function/transform to apply to the images (e.g., resize, normalization, etc.).
            text_processor (transformer): Pre-trained text processor (e.g., CLIP tokenizer).
            max_text_length (int): Maximum length of the tokenized text.
            negative_sample_prob (float): Probability of returning a negative sample.
        """
        self.data = dataset
        self.transform = transform
        self.text_processor = text_processor
        self.max_text_length = max_text_length
        self.negative_sample_prob = negative_sample_prob

    def __len__(self):
        """Return the total number of samples in the dataset."""
        return len(self.data)

    def __getitem__(self, idx):
        """Fetch one sample of image-text pair from the dataset."""
        row = self.data[idx]

        try:
            # Extract image and caption
            query_image = row['image']  # Directly access the 'image' field
            if isinstance(query_image, Image.Image):
                query_image = query_image.convert("RGB")  # Ensure image is in RGB mode

            if self.transform:
                query_image = self.transform(query_image)  # Apply any required transformations

            # Extract caption
            caption = row['caption']

            # Tokenize text using the text processor (e.g., CLIP tokenizer)
            text_inputs = self.text_processor(
                text=caption,
                padding='max_length',
                truncation=True,
                max_length=self.max_text_length,
                return_tensors="pt"
            )

            # Optional: Sample a negative image with some probability
            negative_image = None
            if random.random() < self.negative_sample_prob:
                negative_image = self.get_negative_sample(idx)

            return query_image, text_inputs, row['id'], negative_image

        except Exception as e:
            print(f"Error at index {idx}: {e}")
            return self.get_dummy_sample()  # Return a dummy sample if something goes wrong

    def get_negative_sample(self, idx):
        """Randomly sample a negative image that is not the current query."""
        negative_idx = random.randint(0, len(self.data) - 1)

        # Ensure that we don't pick the same image as the positive one
        while negative_idx == idx:
            negative_idx = random.randint(0, len(self.data) - 1)

        try:
            row = self.data[negative_idx]
            negative_image = row['image']
            if isinstance(negative_image, Image.Image):
                negative_image = negative_image.convert("RGB")

            if self.transform:
                negative_image = self.transform(negative_image)

            return negative_image

        except Exception as e:
            print(f"Error loading negative sample at index {negative_idx}: {e}")
            return None  # Handle any error loading the negative image

    def get_dummy_sample(self):
        """Return a dummy sample in case of invalid data."""
        dummy_image = np.zeros((3, 224, 224), dtype=np.float32)  # A dummy black image
        dummy_text = {
            'input_ids': torch.tensor([[0]]),
            'attention_mask': torch.tensor([[0]])
        }
        return torch.tensor(dummy_image), dummy_text, 'dummy_id', None

```

```python
class Combiner(nn.Module):
    """
    Combiner module which once trained fuses textual and visual information
    """

    def __init__(self, clip_feature_dim=512, projection_dim=512, hidden_dim=512):
        """
        :param clip_feature_dim: CLIP input feature dimension
        :param projection_dim: projection dimension
        :param hidden_dim: hidden dimension
        """
        super(Combiner, self).__init__()
        self.text_projection_layer = nn.Linear(clip_feature_dim, projection_dim)
        self.image_projection_layer = nn.Linear(clip_feature_dim, projection_dim)

        self.dropout1 = nn.Dropout(0.5)
        self.dropout2 = nn.Dropout(0.5)

        self.combiner_layer = nn.Linear(projection_dim * 2, hidden_dim)
        self.output_layer = nn.Linear(hidden_dim, clip_feature_dim)

        self.dropout3 = nn.Dropout(0.5)
        self.dynamic_scalar = nn.Sequential(nn.Linear(projection_dim * 2, hidden_dim), nn.ReLU(), nn.Dropout(0.5),
                                            nn.Linear(hidden_dim, 1), nn.Sigmoid())

        self.logit_scale = 100

    def forward(self, image_features: torch.tensor, text_features: torch.tensor,
                target_features: torch.tensor) -> torch.tensor:
        """
        Takes as input a triplet: image_features, text_features and target_features and outputs the logits which are
        the normalized dot product between the predicted features and the target_features.
        The logits are also multiplied by logit_scale parameter
        :param image_features: CLIP reference image features
        :param text_features: CLIP relative caption features
        :param target_features: CLIP target image features
        :return: scaled logits
        """
        predicted_features = self.combine_features(image_features, text_features)
        target_features = F.normalize(target_features, dim=-1)

        logits = self.logit_scale * predicted_features @ target_features.T
        return logits

    def combine_features(self, image_features: torch.tensor, text_features: torch.tensor) -> torch.tensor:
        """
        Combine the reference image features and the caption features. It outputs the predicted features
        :param image_features: CLIP reference image features
        :param text_features: CLIP relative caption features
        :return: predicted features
        """
        text_projected_features = self.dropout1(F.relu(self.text_projection_layer(text_features)))
        image_projected_features = self.dropout2(F.relu(self.image_projection_layer(image_features)))

        raw_combined_features = torch.cat((text_projected_features, image_projected_features), -1)
        combined_features = self.dropout3(F.relu(self.combiner_layer(raw_combined_features)))
        dynamic_scalar = self.dynamic_scalar(raw_combined_features)
        output = self.output_layer(combined_features) + dynamic_scalar * text_features + (
                1 - dynamic_scalar) * image_features
        return F.normalize(output, dim=-1)
```

```python
import torch
import torch.nn as nn
from transformers import CLIPModel
import torch.nn.functional as F
from torch.cuda.amp import autocast, GradScaler




class AdvancedAdapter(nn.Module):
    """
    A more advanced Adapter Layer with LayerNorm, Dropout, and Residual Connection.
    """
    def __init__(self, input_dim, hidden_dim=512, dropout=0.2):
        super(AdvancedAdapter, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.layernorm = nn.LayerNorm(hidden_dim)
        self.activation = nn.ReLU()
        self.fc2 = nn.Linear(hidden_dim, input_dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        """
        Forward pass with a residual connection.
        """
        residual = x  # Store the original input for the skip connection
        x = self.fc1(x)
        x = self.activation(x)
        x = self.layernorm(x)
        x = self.fc2(x)
        x = self.dropout(x)
        return x + residual  # Residual connection


class ImageTextRetrievalModel(nn.Module):
    """
    Enhanced Retrieval Model based on CLIP with advanced Adapters and improved feature extraction.
    """
    def __init__(self, freeze_clip=True, hidden_dim=512, dropout=0.2):
        super(ImageTextRetrievalModel, self).__init__()

        # Load the base CLIP model
        self.clip_model = CLIPModel.from_pretrained('openai/clip-vit-large-patch14')

        # Add advanced Adapters to text and vision encoders
        for i in range(len(self.clip_model.text_model.encoder.layers)):
            adapter = AdvancedAdapter(input_dim=768, hidden_dim=hidden_dim, dropout=dropout)
            self.clip_model.text_model.encoder.layers[i].add_module('adapter', adapter)

        for i in range(len(self.clip_model.vision_model.encoder.layers)):
            adapter = AdvancedAdapter(input_dim=768, hidden_dim=hidden_dim, dropout=dropout)
            self.clip_model.vision_model.encoder.layers[i].add_module('adapter', adapter)

        self.freeze_clip = freeze_clip
        if self.freeze_clip:
            self.freeze_clip_model()

        # Unfreeze the last two layers of the text and vision encoders
        self.unfreeze_last_two_layers()

    def freeze_clip_model(self):
        """Freeze all CLIP layers to prevent fine-tuning (save memory and prevent overfitting)."""
        for param in self.clip_model.parameters():
            param.requires_grad = False

    def unfreeze_last_two_layers(self):
        """Unfreeze the last two layers of both text and vision encoders."""
        # Unfreeze the last two layers of the text encoder
        for i in range(2):
            for param in self.clip_model.text_model.encoder.layers[-(i+1)].parameters():
                param.requires_grad = True

        # Unfreeze the last two layers of the vision encoder
        for i in range(2):
            for param in self.clip_model.vision_model.encoder.layers[-(i+1)].parameters():
                param.requires_grad = True

    def forward(self, image, text_inputs):
        """Extract features from image and text."""
        # Extract image features with the vision encoder
        image_features = self.clip_model.get_image_features(image)

        # Extract text features with the text encoder
        text_features = self.clip_model.get_text_features(**text_inputs)

        return image_features, text_features

    def compute_similarity(self, image_features, text_features):
        """
        Computes cosine similarity between image and text features for retrieval.
        """
        # Normalize the features to unit length for cosine similarity
        image_features = F.normalize(image_features, p=2, dim=-1)
        text_features = F.normalize(text_features, p=2, dim=-1)

        # Compute cosine similarity
        similarity = torch.matmul(image_features, text_features.T)
        return similarity

    def contrastive_loss(self, image_features, text_features, temperature=0.07):
        """
        Computes a contrastive loss (using InfoNCE) between image and text features.
        """
        # Normalize features
        image_features = F.normalize(image_features, p=2, dim=-1)
        text_features = F.normalize(text_features, p=2, dim=-1)

        # Compute similarity matrix (cosine similarity)
        similarity_matrix = torch.matmul(image_features, text_features.T)  # Shape: (batch_size, batch_size)

        # Create labels for contrastive loss (positive pair diagonal is 1)
        labels = torch.arange(image_features.size(0)).to(image_features.device)

        # Loss computation (InfoNCE loss)
        logits = similarity_matrix / temperature
        loss = F.cross_entropy(logits, labels)

        return loss
```

```python
class ContrastiveLoss(nn.Module):
    def __init__(self, margin=1.0, temperature=0.07):
        """
        Implements a contrastive loss using dot-product similarity with temperature scaling,
        hard-negative mining, and batch-wise InfoNCE loss.

        Args:
        - margin (float): Margin for ranking loss (only used for ranking loss logic).
        - temperature (float): Temperature to scale the dot product similarities.
        """
        super(ContrastiveLoss, self).__init__()
        self.margin = margin
        self.temperature = temperature
        self.loss_fn = nn.MarginRankingLoss(margin=self.margin)

    def forward(self, image_features, text_features, target_image_features=None):
        """
        Calculate the contrastive loss for image-text retrieval.

        Args:
        - image_features: Tensor of shape (batch_size, feature_dim) for query image features.
        - text_features: Tensor of shape (batch_size, feature_dim) for text features (positive pair with query image).
        - target_image_features: Tensor of shape (batch_size, feature_dim) for target (negative) image features.

        Returns:
        - loss: Calculated contrastive loss for the batch.
        """
        # Ensure the batch sizes match
        batch_size = image_features.size(0)
        assert image_features.size(0) == text_features.size(0), \
            "Batch sizes for image and text features must be the same"

        if target_image_features is not None:
            assert target_image_features.size(0) == batch_size, \
                "Batch sizes for image, text, and target image features must be the same"

        # **1. Batch-Wise InfoNCE Loss**
        # Calculate pairwise dot product similarity between all image features and text features
        similarity_matrix = torch.matmul(image_features, text_features.T)  # Shape (batch_size, batch_size)
        similarity_matrix /= self.temperature  # Apply temperature scaling

        # Labels for InfoNCE (positive pairs are on the diagonal)
        labels = torch.arange(batch_size).to(image_features.device)  # Ground truth labels for positives

        # Calculate the cross-entropy loss using similarity matrix
        info_nce_loss = F.cross_entropy(similarity_matrix, labels)

        # **2. Hard-Negative Mining**
        if target_image_features is not None:
            # Calculate similarity between image and all other batch images
            hard_negatives_matrix = torch.matmul(image_features, target_image_features.T)  # (batch_size, batch_size)
            hard_negatives_matrix.fill_diagonal_(-float('inf'))  # Remove self-matching
            hardest_neg_index = torch.argmax(hard_negatives_matrix, dim=-1)  # Get hardest negative index for each image
            hardest_neg_features = target_image_features[hardest_neg_index]  # Hardest negatives for each image

            # Calculate similarity between image and hardest negative
            similarity_pos = torch.sum(image_features * text_features, dim=-1)  # Positive pair similarity
            similarity_hard_neg = torch.sum(image_features * hardest_neg_features, dim=-1)  # Hard negative similarity

            # Calculate the ranking loss using MarginRankingLoss
            target = torch.ones_like(similarity_pos)  # Target is 1 since positive should be greater than negative
            hard_negative_loss = self.loss_fn(similarity_pos, similarity_hard_neg, target)
        else:
            hard_negative_loss = 0.0

        # **3. Total Loss**
        total_loss = info_nce_loss + hard_negative_loss  # Combine InfoNCE loss with hard-negative mining loss

        return total_loss, info_nce_loss, hard_negative_loss
```

```python
def prepare_dataloaders(batch_size=32, dataset_names=['mscoco', 'flickr30k', 'cirr'], cache_dir='./cache'):
    """
    Load and prepare the DataLoader for training and validation.

    Args:
    - batch_size (int): Batch size for training and validation.
    - dataset_names (list): List of datasets to load (e.g., 'mscoco', 'flickr30k', 'cirr').
    - cache_dir (str): Directory to cache the dataset files.

    Returns:
    - train_dataloader (DataLoader): DataLoader for training.
    - val_dataloader (DataLoader): DataLoader for validation.
    """

    # **1. Define Image Transformations**
    train_transform = transforms.Compose([
        transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),  # Random crop
        transforms.RandomHorizontalFlip(p=0.5),  # Random horizontal flip
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),  # Random color jitter
        transforms.RandomRotation(degrees=10),  # Small rotation
        transforms.ToTensor(),
        transforms.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225))  # ImageNet normalization
    ])

    val_transform = transforms.Compose([
        transforms.Resize(256),  # Resize shortest side to 256
        transforms.CenterCrop(224),  # Center crop to 224x224
        transforms.ToTensor(),
        transforms.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225))  # ImageNet normalization
    ])

    # **2. Load Datasets**
    datasets = {}

    if 'cirr' in dataset_names:
        datasets['cirr'] = load_dataset("royokong/cirr_val", cache_dir=cache_dir)

    if 'mscoco' in dataset_names:
        datasets['mscoco'] = load_dataset("nlphuji/mscoco_2014_5k_test_image_text_retrieval", cache_dir=cache_dir)

    if 'flickr30k' in dataset_names:
        datasets['flickr30k'] = load_dataset("nlphuji/flickr30k", cache_dir=cache_dir)

    # **3. Combine and Format Datasets**
    train_datasets = []
    val_datasets = []

    for dataset_name, dataset_dict in datasets.items():
        # If a dataset has a 'train' split, add it to train_datasets.
        # Otherwise, add 'val' or 'test' split to train_datasets as a substitute.
        if 'train' in dataset_dict:
            train_datasets.append(dataset_dict['train'])
        else: # Added this else block to use 'val' or 'test' for training
            if 'val' in dataset_dict:
                train_datasets.append(dataset_dict['val'])
                print(f"Using 'val' split of {dataset_name} for training.")
            elif 'test' in dataset_dict:
                train_datasets.append(dataset_dict['test'])
                print(f"Using 'test' split of {dataset_name} for training.")

        # Still add validation/test splits to val_datasets
        if 'val' in dataset_dict:
            val_datasets.append(dataset_dict['val'])
        elif 'test' in dataset_dict:  # Use test set as validation if no val split exists
            val_datasets.append(dataset_dict['test'])

    # **4. Wrap Each Dataset in a Custom RetrievalDataset**
    processor = CLIPProcessor.from_pretrained('openai/clip-vit-base-patch32')

    train_dataset = torch.utils.data.ConcatDataset([
        RetrievalDataset(ds, transform=train_transform, text_processor=processor.tokenizer, max_text_length=77)
        for ds in train_datasets
    ])

    val_dataset = torch.utils.data.ConcatDataset([
        RetrievalDataset(ds, transform=val_transform, text_processor=processor.tokenizer, max_text_length=77)
        for ds in val_datasets
    ])

    # **5. Create DataLoader with Collate Function**
    train_dataloader = DataLoader(
        train_dataset,
        batch_size=batch_size,
        shuffle=True,
        collate_fn=collate_fn,
        pin_memory=True,
        num_workers=4
    )

    val_dataloader = DataLoader(
        val_dataset,
        batch_size=batch_size,
        shuffle=False,
        collate_fn=collate_fn,
        pin_memory=True,
        num_workers=4
    )

    return train_dataloader, val_dataloader
```

```python
def train_model(model, train_loader, val_loader, epochs, learning_rate, device):
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    criterion = ContrastiveLoss()
    model.to(device)

    for epoch in range(epochs):
        model.train()
        total_loss = 0.0
        for batch in train_loader:
            images, text_inputs, ids, neg_images = batch
            images, neg_images = images.to(device), (neg_images.to(device) if neg_images is not None else None)
            text_inputs = {k: v.to(device) for k, v in text_inputs.items()}

            optimizer.zero_grad()
            image_features, text_features = model(images, text_inputs)
            neg_features = model.clip_model.get_image_features(neg_images) if neg_images is not None else None
            loss, info_nce_loss, hard_neg_loss = criterion(image_features, text_features, neg_features)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

        print(f"Epoch [{epoch+1}/{epochs}], Loss: {total_loss / len(train_loader)}")

        model.eval()
        with torch.no_grad():
            val_loss = 0.0
            for batch in val_loader:
                images, text_inputs, ids, neg_images = batch
                images = images.to(device)
                text_inputs = {k: v.to(device) for k, v in text_inputs.items()}
                image_features, text_features = model(images, text_inputs)
                neg_features = model.clip_model.get_image_features(neg_images) if neg_images is not None else None
                loss, _, _ = criterion(image_features, text_features, neg_features)
                val_loss += loss.item()
            print(f"Validation Loss: {val_loss / len(val_loader)}")
```

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
import torch.optim as optim

def train(model, dataloader, criterion, optimizer, device, epochs=2, accumulation_steps=4):
    """
    Train the model with gradient accumulation for larger effective batch sizes.

    Arguments:
    - model: The model to be trained (ImageTextRetrievalModel).
    - dataloader: DataLoader providing training batches.
    - criterion: Loss function to optimize (ContrastiveLoss).
    - optimizer: Optimizer for model parameters.
    - device: The device (e.g., 'cuda' or 'cpu') where the model and data should be moved.
    - epochs: Number of epochs for training.
    - accumulation_steps: Number of steps to accumulate gradients before updating weights.
    """

    model.to(device)  # Move the model to the correct device
    model.train()  # Set the model to training mode

    for epoch in range(epochs):
        total_loss = 0
        optimizer.zero_grad()  # Zero the gradients at the beginning of each epoch

        for batch_idx, (query_image, target_image, text_inputs) in enumerate(dataloader): # Changed order of unpacking
            # Move data to the appropriate device (GPU or CPU)
            query_image = query_image.to(device)
            target_image = target_image.to(device)
            text_inputs = {k: v.to(device) for k, v in text_inputs.items()}  # Move text inputs to device

            # Forward pass through the model
            # Enable gradients for image tensors
            query_image.requires_grad = True
            target_image.requires_grad = True

            image_features, text_features = model(query_image, text_inputs)
            target_image_features, _ = model(target_image, text_inputs)

            # Calculate similarity (e.g., using cosine similarity or triplet loss)
            loss = criterion(image_features, text_features, target_image_features)

            loss = loss / accumulation_steps  # Normalize the loss for gradient accumulation
            loss.backward()  # Backpropagate the gradients

            if (batch_idx + 1) % accumulation_steps == 0:  # Update weights every accumulation_steps
                optimizer.step()  # Update model weights
                optimizer.zero_grad()  # Reset gradients

            total_loss += loss.item()

            # Optionally, print the loss after every few batches for progress monitoring
            if (batch_idx + 1) % 10 == 0:
                print(f"Epoch [{epoch+1}/{epochs}], Batch [{batch_idx+1}/{len(dataloader)}], Loss: {loss.item():.4f}")

        # Print average loss at the end of the epoch
        print(f"Epoch [{epoch+1}/{epochs}], Average Loss: {total_loss / len(dataloader):.4f}")

    # Save model weights after training completes
    torch.save(model.state_dict(), 'weights.pth')
    print("Training complete. Model weights saved to 'weights.pth'")

# Example Usage
if __name__ == "__main__":
    # Initialize model, criterion, optimizer, and DataLoader
    device = 'cuda' if torch.cuda.is_available() else 'cpu'

    model = ImageTextRetrievalModel()  # Your model class (ImageTextRetrievalModel)
    model.to(device)  # Move model to device

    criterion = ContrastiveLoss()  # Your custom loss function (e.g., ContrastiveLoss)
    optimizer = optim.Adam(model.parameters(), lr=1e-3)  # Optimizer with a learning rate of 1e-4
    dataset = RetrievalDataset(csv_file="/content/dataset_folder/train_data.csv", image_folder="/content/dataset_folder/images",
                           transform=image_transform_train, text_processor=clip_processor)
    # Initialize the DataLoader (assuming your dataset and collate_fn are already defined)
    dataloader = DataLoader(dataset, batch_size=4, shuffle=True, collate_fn=collate_fn, pin_memory=True)

    # Start training the model
    train(model, dataloader, criterion, optimizer, device, epochs=1, accumulation_steps=4)

```

```python
# from model import ImageTextRetrievalModel
def load_model(weights_path='weights.pth'):
    """Load the trained model with pre-trained weights."""
    model = ImageTextRetrievalModel().to(device)
    model.load_state_dict(torch.load(weights_path, map_location=torch.device('cuda' if torch.cuda.is_available() else 'cpu')))
    model.eval()
    return model
```

```python
def encode_queries(df: pd.DataFrame) -> np.ndarray:
    """
    Process query pairs and generate embeddings using the pre-trained model.

    Args:
        df (pd.DataFrame): DataFrame with columns:
            - query_image: str, paths to query images
            - query_text: str, text descriptions

    Returns:
        np.ndarray: Embeddings array (num_queries, embedding_dim)
    """
    model = load_model()  # Load the model once
    query_embeddings = []

    # Process each query and generate embeddings
    for idx, query in df.iterrows():
        query_image_path = query['query_image']
        query_text = query['query_text']

        # Generate query embedding using the model
        query_embedding = get_query_embedding(query_image_path, query_text, model)
        query_embeddings.append(query_embedding)

    return np.vstack(query_embeddings)
```

```python
def get_query_embedding(query_image_path, query_text, model):
    """
    Generate embedding for a query image and text.

    Args:
        query_image_path (str): Path to the query image.
        query_text (str): The query text.
        model (torch.nn.Module): Pre-trained model.

    Returns:
        np.ndarray: Query embedding (1, embedding_dim)
    """
    query_image_path = f"/content/dataset_folder/images/{query_image_path}"
    processor = CLIPProcessor.from_pretrained('openai/clip-vit-base-patch32')
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])

    image = Image.open(query_image_path).convert('RGB')
    image = transform(image).unsqueeze(0).to(device)

    text_inputs = processor(text=[query_text], return_tensors="pt", padding=True, truncation=True, max_length=32)
    text_inputs = {k: v.to(device) for k, v in text_inputs.items()}

    with torch.no_grad():
        image_features, text_features = model(image, text_inputs)
        query_embedding = (image_features+ text_features)

    return query_embedding.cpu().numpy()
```

```python
def encode_database(df: pd.DataFrame) -> np.ndarray:
    """
    Process database images and generate embeddings using the pre-trained model.

    Args:
        df (pd.DataFrame): DataFrame with column:
            - target_image: str, paths to database images

    Returns:
        np.ndarray: Embeddings array (num_images, embedding_dim)
    """
    model = load_model()  # Load the model once
    database_embeddings = []

    # Process each image in the database
    for idx, row in df.iterrows():
        image_path = row['target_image']

        # Generate database image embedding using the model
        image_embedding = get_database_embedding(image_path, model)
        database_embeddings.append(image_embedding)

    return np.vstack(database_embeddings)

```

```python
def get_database_embedding(image_path, model):
    """
    Generate embedding for a database image.

    Args:
        image_path (str): Path to the image.
        model (torch.nn.Module): Pre-trained model.

    Returns:
        np.ndarray: Image embedding (1, embedding_dim)
    """
    image_path = f"/content/dataset_folder/images/{image_path}"
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])

    image = Image.open(image_path).convert('RGB')
    image = transform(image).unsqueeze(0).to(device)

    with torch.no_grad():
        image_features = model.clip_model.get_image_features(image)

    return image_features.cpu().numpy()
```
